```python
#Creating the for loop test
new_bel = []
new_lux = []

percentage_bel_lst = []
percentage_lux_lst = []

for bel, lux in zip(Belgium, Luxembourg):
    
    
    if bel != 0 and lux !=0:
        
        bel_percentage = (bel/(bel+lux))
        percentage_bel_lst.append(bel_percentage)
        lux_percentage = (lux/(bel+lux))
        percentage_lux_lst.append(lux_percentage)

for belux in Bel_Lux: 
    if belux != 0:
        new_bel.append(int(belux*np.mean(percentage_bel_lst)))
        new_lux.append(int(belux*np.mean(percentage_lux_lst)))
        
```
